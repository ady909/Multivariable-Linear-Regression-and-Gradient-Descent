```python
#importing libraries
from sklearn.datasets import fetch_california_housing
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading dataset
housing = fetch_california_housing()
```

```python
#dataset insights
housing
```

```python
#Splitting data into 2 parts
X = housing.data
Y = housing.target
X.shape , Y.shape
```

```python
#dataframe formation
df_housing = pd.DataFrame(X)
df_housing
```

```python
#adding column names to df
df_housing.columns = housing.feature_names
df_housing
```

```python
#adding new column
df_housing["HouseAge_	HouseAge"] = df_housing.HouseAge ** 2
X2 = df_housing.values
X2.shape
```

```python
df_housing
```

```python
#splitting data
from sklearn import model_selection
X_train ,X_test ,Y_train ,Y_test = model_selection.train_test_split(X , Y , random_state = 0)
X2_train ,X2_test ,Y2_train ,Y2_test = model_selection.train_test_split(X2 , Y)
```

```python
#creating instances of linear regression algorithm
from sklearn.linear_model import LinearRegression
alg1 = LinearRegression()
alg2 = LinearRegression()
```

```python
#fitting our training data inside algorithm
alg1.fit(X_train , Y2_train)
alg2.fit(X2_train , Y2_train)
```

```python
#finding score for Algorithm 1
Y_pred = alg1.predict(X_test)
training_score = alg1.score(X_train ,Y_train)
testing_score = alg1.score(X_test , Y_test)
print(f"Training Data score: {training_score}")
print(f"Testing Data score: {testing_score}")

#finding score for Algorithm 2
training_score = alg2.score(X2_train ,Y2_train)
testing_score = alg2.score(X2_test , Y2_test)
print(f"Training Data score: {training_score}")
print(f"Testing Data score: {testing_score}")
```
