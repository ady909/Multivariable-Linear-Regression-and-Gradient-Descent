```python
def step_gradient(points,learning_rate,m,c):
  m_slope = 0
  c_slope = 0
  for i in range(len(points)):
    x = points[i , 0]
    y = points[i,1]
    m_slope += ((-2/ len(points) * ((y-m*x-c))*x) )
    c_slope += ((-2/ len(points) * ((y-m*x-c))) )
  new_m = m - learning_rate * m_slope
  new_c = c - learning_rate * c_slope
  return new_m,new_c
```

```python
def gd(points,learning_rate,no_iterations):
  m = 0
  c = 0
  for i in range(len(points)):
    m,c = step_gradient(points,learning_rate,m,c)
    print(i , f"cost:{cost(points,m,c)}")
  return m,c
```

```python
def cost(points,m,c):
  total_cost = 0
  for i in range(len(points)):
    x = points[i,0]
    y = points[i,1]
    total_cost += ((y - m*x -c)**2 ) / len(points)
  return total_cost
```

```python
def run():
  data = np.loadtxt("data.csv" , delimiter = ",")
  learning_rate = 0.0001
  no_iterations = 10
  m,c = gd(data,learning_rate,no_iterations)
  print(m,c)
```
